Go to Docker Hub and search for the rocker/rstudio image.
Ensure the image supports both linux/amd64 and linux/arm64 architectures.

Open your terminal or command prompt.

Task 4: Create a 2x2 Multiplot with Various Charts
In RStudio, use the following R code to create a 2x2 multiplot with a histogram, a boxplot, a barchart, and a piechart using arbitrary data:
r
Code kopieren
# Set up the plotting area
par(mfrow=c(2,2))

# Histogram
hist(rnorm(100), main="Histogram", col="blue", border="black")

# Boxplot
boxplot(rnorm(100), main="Boxplot", col="red")

# Barchart
barplot(table(sample(letters[1:5], 100, replace=TRUE)), main="Barchart", col="green")

# Piechart
pie(table(sample(letters[1:4], 10, replace=TRUE)), main="Piechart", col=rainbow(4))
Task 5: Save the Resulting Plots as .png Files
Use the following R code to save each plot as a separate PNG file:
r
Code kopieren
# Save histogram
png("histogram.png")
hist(rnorm(100), main="Histogram", col="blue", border="black")
dev.off()

# Save boxplot
png("boxplot.png")
boxplot(rnorm(100), main="Boxplot", col="red")
dev.off()

# Save barchart
png("barchart.png")
barplot(table(sample(letters[1:5], 100, replace=TRUE)), main="Barchart", col="green")
dev.off()

# Save piechart
png("piechart.png")
pie(table(sample(letters[1:4], 10, replace=TRUE)), main="Piechart", col=rainbow(4))
dev.off()
After running these commands, you should find the resulting .png files (histogram.png, boxplot.png, barchart.png, piechart.png) saved in your working directory.
