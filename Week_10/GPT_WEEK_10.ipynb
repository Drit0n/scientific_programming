Erstellung von Grafiken mit Matplotlib und Seaborn.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinuskurve')
plt.legend()
plt.show()
```
